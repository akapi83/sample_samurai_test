```python
import pandas as pd



```

```python
import os

def load_order_excel(filename):
  """指定されたファイル名で`order_new/`ディレクトリからExcelファイルを読み込む関数"""
  file_path = os.path.join('/content/order_new', filename)
  try:
    df = pd.read_excel(file_path)
    print(f"{filename} を正常に読み込みました。")
    return df
  except FileNotFoundError:
    print(f"エラー: {file_path} が見つかりません。")
    return None
  except Exception as e:
    print(f"エラー: {filename} の読み込み中に問題が発生しました: {e}")
    return None

load_order_excel('order_A_20230524.xlsx')
```

### `order_new`ディレクトリ内のすべてのExcelファイルを読み込む

```python
import os

# `order_new`ディレクトリのパス
order_new_dir = '/content/order_new'

# ディレクトリ内のすべてのファイルを取得
all_files = os.listdir(order_new_dir)

# Excelファイルのみをフィルタリング
excel_files = [f for f in all_files if f.endswith(('.xlsx', '.xls'))]

# 読み込んだDataFrameを格納する辞書
order_dfs = {}

print(f"以下のExcelファイルを読み込みます: {excel_files}")

for filename in excel_files:
  df = load_order_excel(filename)
  if df is not None:
    # ファイル名から拡張子を除いたものをキーとして使用
    key_name = os.path.splitext(filename)[0]
    order_dfs[key_name] = df
    print(f"\n--- {filename} の最初の5行 ---")
    display(df.head())
```

すべてのDataFrameが`order_dfs`辞書に格納されました。辞書のキーはファイル名（拡張子なし）です。
